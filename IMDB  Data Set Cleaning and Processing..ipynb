```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_actual=pd.read_csv('imdb.csv')
```

```python
df_actual.info()
```

```python
df_actual.drop(df_actual[df_actual['Unnamed: 46'].notnull()].index,inplace=True)
```

```python
df_actual.drop(df_actual[df_actual['Unnamed: 47'].notnull()].index,inplace=True)
```

```python
#making of df1 and removing df1 from df_actual
df_1=df_actual[df_actual['Extra1'].notnull()]
df_actual.drop(df_1.index,inplace=True)
```

```python
#making of df_2 and removing df2 from df1
df_2=df_1[df_1['Extra2'].notnull()]
df_1.drop(df_2.index, inplace=True)
```

```python
df_2.head()
```

```python
#now adding title, wordsintitle and url and giving the value to url, so that 
#when we shift the dataframe we get thr values directly under title.

df_2['url']=df_2['title']+df_2['wordsInTitle']+df_2['url']
```

```python
df_2.head()
```

```python
#now shifting the data towards left side upto title.
#spliting the dataframe into tow parts, 
#performing shifting operations on second part
#concatinating both the dataFrames on axis=1
df_2_1=df_2[['fn','tid']]
df_2_2=df_2.drop(columns=['fn','tid'])
```

```python
#now performing the shifting operation on df_2_2:
#periods means no.of times you have to shift.
#shifting towards left so giving axis 1 and -2 because for left - is used

df_2_2=df_2_2.shift(periods=-2,axis=1)
```

```python
#again joining df_2_1 and df_2_2 and storing in df_2
df_2=pd.concat([df_2_1,df_2_2],axis=1)
```

```python
#performing the same operations now on df_1:
df_1_1=df_1[['fn','tid']]
df_1_2=df_1.drop(columns=['fn','tid'])
```

```python
#now performing the shifting operation on df_1_2:
#periods means no.of times you have to shift.
#shifting towards left so giving axis 1 and -1 because for left - is used
df_1_2=df_1_2.shift(periods=-1,axis=1)
```

```python
#Again joining df_1_1 and df_1_2
df_1=pd.concat([df_1_1,df_1_2],axis=1)
```

```python
# Now concatinating df_actual,df_1 and df_2 into new dataframe df:
df=pd.concat([df_actual,df_1,df_2],axis=0)
```

```python
df.info()
```

```python
#now we will perfrom our processing on df.
columns_d=['Extra1','Extra2','Unnamed: 46','Unnamed: 47']
df.drop(columns=columns_d,inplace=True)
```

```python
df[['imdbRating','duration','year','nrOfWins','nrOfNominations','nrOfPhotos','nrOfNewsArticles','nrOfUserReviews']] = df[['imdbRating','duration','year','nrOfWins','nrOfNominations','nrOfPhotos','nrOfNewsArticles','nrOfUserReviews']].astype('float64')
```

### Solution 6:

```python
#q6: find top 5 genre with hightest no of user reviews:
  #[::-1]
    
####### METHOD 1::::: using loop:
genres=['Action', 'Adult', 'Adventure', 'Animation', 'Biography',
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy',
       'FilmNoir', 'GameShow', 'History', 'Horror', 'Music', 'Musical',
       'Mystery', 'News', 'RealityTV', 'Romance', 'SciFi', 'Short', 'Sport',
       'TalkShow', 'Thriller', 'War', 'Western']
d={}
for g in genres:
    d[g]=df.groupby([g])['nrOfUserReviews'].sum()[1]
#d_df=pd.DataFrame(d)


####### METHOD 2::::: using decoding: doing reverse of one hot encoding:
#jis column m 1 hai us column ka nam mil jaye uske liye:
genres=['Action', 'Adult', 'Adventure', 'Animation', 'Biography',
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy',
       'FilmNoir', 'GameShow', 'History', 'Horror', 'Music', 'Musical',
       'Mystery', 'News', 'RealityTV', 'Romance', 'SciFi', 'Short', 'Sport',
       'TalkShow', 'Thriller', 'War', 'Western']

df_g=df[genres]

#defining function , it will work on each column and will check each row
#where 1 is.


def get_genre(row):
    for g in genres:
        if row[g] == 1.0:
            return g
    
df['genres']=df_g.apply(get_genre,axis=1)

#ye har column ki har row pe jake check krega.
```

```python
df.groupby('genres')['nrOfUserReviews'].sum().sort_values(ascending=False).head()

#for getting output in dataframe:
df.groupby('genres')[['genres','nrOfUserReviews']].sum().sort_values(by='nrOfUserReviews',ascending=False).head()
```
